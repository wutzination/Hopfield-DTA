```python
#Get early stopping point
def early_stopping(filename:str):
  #print(filename)
  if not os.path.isfile(filename):
    print(f"{filename} is a directory")
    return
  
  df = pd.read_csv(filename)
  df.columns = ['epoch', 'step', 'step_min', 'step_max', 'mean', 'min', 'max']
  if "mse" in filename:
    epoch = np.argmin(df["mean"])
    value_type = 'mse'
  elif "ci" in filename:
    epoch = np.argmax(df["mean"])
    value_type = 'ci'
  else:
    print("Invalid File Name! mse or ci missing.")
    return 

  value = df["mean"][epoch]

  #print(f'Stopping Point for {value_type} at epoch {epoch} with value {value}')
  return {"filename":filename, "value_type":value_type, "epoch":epoch, "value":value}
```

```python
file_list = sorted(os.listdir('./'))
```

```python
#Save results to Dataframe
result_df = pd.DataFrame(columns=["filename", "value_type", "epoch", "value"])
#loop through files
for f in file_list:
  result = early_stopping(f)
  #print(result)
  result_df=result_df.append(result, ignore_index=True)
```
