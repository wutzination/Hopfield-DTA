```python
#paths
path_train_fold1 = 'train_fold1_KIBA.txt'
path_train_fold2 = 'train_fold2_KIBA.txt'
path_train_fold3 = 'train_fold3_KIBA.txt'
path_train_fold4 = 'train_fold4_KIBA.txt'
path_train_fold5 = 'train_fold5_KIBA.txt'

save_split1_train = 'split1/train1.txt'
save_split1_val =  'split1/val1.txt'
save_split2_train = 'split2/train2.txt'
save_split2_val =  'split2/val2.txt'
save_split3_train = 'split3/train3.txt'
save_split3_val =  'split3/val3.txt'
save_split4_train = 'split4/train4.txt'
save_split4_val =  'split4/val4.txt'
save_split5_train = 'split5/train5.txt'
save_split5_val =  'split5/val5.txt'
```

```python
#load features
def get_feature_list(inpath:str):
  feature_list = []
  with open(inpath) as f:
    for line in f:
      feature_list.append(line)
  return feature_list

#folds drugs
train_fold1=get_feature_list(path_train_fold1)
train_fold2=get_feature_list(path_train_fold2)
train_fold3=get_feature_list(path_train_fold3)
train_fold4=get_feature_list(path_train_fold4)
train_fold5=get_feature_list(path_train_fold5)
```

```python
#Split1 
split1_train = train_fold1+train_fold2 +train_fold3+train_fold4
split1_val= train_fold5
print(f'Len Train: \t{len(split1_train)}')
print(f'Len Val: \t{len(split1_val)}')
with open(save_split1_train, "w") as fp:  
  for line in split1_train:
    fp.write(line)
with open(save_split1_val, "w") as fp:  
  for line in split1_val:
    fp.write(line)
```

```python
#Split2 
split2_train = train_fold1+train_fold2 +train_fold3+train_fold5
split2_val= train_fold4
print(f'Len Train: \t{len(split2_train)}')
print(f'Len Val: \t{len(split2_val)}')
with open(save_split2_train, "w") as fp:  
  for line in split2_train:
    fp.write(line)
with open(save_split2_val, "w") as fp:  
  for line in split2_val:
    fp.write(line)
```
